```python
# Import Dependencies
import pandas as pd
import json
import numpy as np
from pathlib import Path
```

```python
# Store filepaths into variable
drivers_csv = Path("data/drivers.csv")
fcyphases_csv = Path("data/fcyphases.csv")
laps_csv = Path("data/laps.csv")
qualifyings_csv = Path("data/qualifyings.csv")
races_csv = Path("data/races.csv")
retirements_csv = Path("data/retirements.csv")
starterfields_csv = Path("data/starterfields.csv")
weather = Path("data/merged_weather.csv")
weather_long = Path("data/f1_weather_long.csv")
```

```python
# Read in files and make dfs
drivers_df = pd.read_csv(drivers_csv)
fcyphases_df = pd.read_csv(fcyphases_csv)
laps_df = pd.read_csv(laps_csv)
qualifyings_df = pd.read_csv(qualifyings_csv)
races_df = pd.read_csv(races_csv)
retirements_df = pd.read_csv(retirements_csv)
starterfields_df = pd.read_csv(starterfields_csv)
weather_df = pd.read_csv(weather)
weather_long_df = pd.read_csv(weather_long)
```

```python
# Rename the 'id' column in drivers_df to 'driver_id'
drivers_df.rename(columns={'id': 'driver_id'}, inplace=True)
races_df.rename(columns={'id': 'race_id'}, inplace=True)
```

```python
mergedDR_df= pd.merge(drivers_df,retirements_df, on="driver_id")
mergedDR_df
```

```python
mergedDRS_df =pd.merge(mergedDR_df,starterfields_df, on="driver_id")
mergedDRS_df
```

```python
mergedDdESRa_df = pd.merge(mergedDRS_df,races_df, on="race_id")
mergedDdESRa_df
```

```python
mergedSRa_df = pd.merge(mergedDRS_df,races_df, on="race_id")
mergedDdESRa_df
```

```python
mergedSRa_df = pd.merge(starterfields_df, races_df, on="race_id")
mergedSRa_df
```

```python
# change "NaN" values in "comments" column to "Null" for read in with json
DNF_data_df = mergedDdESRa_df.fillna("Null")
DNF_data_df
```

```python
DNF_data_df.to_csv("data/dnf/DNF_data.csv", index=False)
mergedDR_df.to_csv("data/dnf/mergedDR.csv", index=False)
mergedSRa_df.to_csv("data/dnf/mergedSRa.csv", index=False)
```

```python
# Change "NaN" values in the "comment" column to "Null"
DNF_data_df['comment'] = DNF_data_df['comment'].fillna('Null')
DNF_data_df['accidents'] = DNF_data_df['accidents'].fillna('Null')
DNF_data_df['failures'] = DNF_data_df['failures'].fillna('Null')
mergedDR_df['accidents'] = mergedDR_df['accidents'].fillna('Null')
mergedDR_df['failures'] = mergedDR_df['failures'].fillna('Null')
mergedSRa_df['comment'] = mergedSRa_df['comment'].fillna('Null')

```

```python
# Now the DataFrame is ready to be saved as a JSON file
# Saving the DataFrame to a JSON file
DNF_data_df.to_json('DNF_data.json', orient='records', lines=True)
mergedDR_df.to_json('mergedDR.json', orient='records', lines=True)
mergedSRa_df.to_json('mergedSRa.json', orient='records', lines=True)
```
